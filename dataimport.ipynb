```python
import pandas as pd 
import pymongo
from bson.objectid import ObjectId
from pprint import pprint

```

```python
#init Mongo
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

#create database
db = client.visualizing_sep
```

```python
#import category tagged csv 

tagged_entries = pd.read_csv('sep.csv',usecols=['_id','Main','Tags'])
tagged_entries.head()
data = list(tagged_entries.to_dict('records'))
for d in data:
    id_val = ObjectId(d['_id'])
    primary_domain = d['Main']
    if type(d['Tags']) == str:
        all_domains = d['Tags']
    else:
        all_domains = ''

    db.sep_data.update_one(
                            {"_id" : id_val}, 
                            {"$set" : 
                                {"primary_domain": primary_domain,
                                 "domain_tags" :all_domains}})
    


```
